```python
# Programa 1: Tienda (invertario) tipo 1
Productos = ["aceite", "pescado", "pan", "papa", "cebolla"]
Cantidad = [3,4,12,10,20]

print(30*"-")
print("Almacenes MercaSuper")
print(30*"-")
print("1.Mostrar productos")
print("2.Compre ya!")
print("3.surtir")

Opcion = int(input())
if Opcion == 1:
  print("Productos:",Productos)
  print("Cantidad:",Cantidad)
elif Opcion == 2:
  print("¿que producto desea?")
  Prodto = input()
  Prodto = Prodto.lower()
  Productos.count(Prodto)
  if Productos.count(Prodto) > 0:
    CodigoDelProducto = Productos.index(Prodto) # posicion = codigo del producto buscado o que se requiere
    print("El codigo del producto",Productos[CodigoDelProducto],"es", CodigoDelProducto,"y tenemos", Cantidad[CodigoDelProducto])
    print("¿cuantos productos desea llevar?")
    Venta = int(input())
    if Venta < Cantidad[CodigoDelProducto]:
      Cantidad[CodigoDelProducto] = Cantidad[CodigoDelProducto] - Venta
      print("usted adquirio",Venta,"nos quedan",Cantidad[CodigoDelProducto]) 
    else:
      print("Disculpenos, no posemos la cantidad requerida, pero tenemos",Cantidad[CodigoDelProducto],"¿los desea adquirir?")
  else:
    print("Disculpe, no tenemo ese producto pero tenemos:", Productos)
elif Opcion == 3:
  print("¿que producto vamos a adquirir?")
  Prodto = input()
  Prodto = Prodto.lower()
  Productos.count(Prodto)
  if Productos.count(Prodto) > 0:
      CodigoDelProducto = Productos.index(Prodto)
      print("¿cuantos productos compraremos?")
      Compra = int(input())
      Cantidad[CodigoDelProducto] = Cantidad[CodigoDelProducto] + Compra
      print(Prodto,"usted adquirio",Compra,"ahora tenemos:",Cantidad[CodigoDelProducto]) 
  else:
    print("¿cuantos productos compraremos?")
    Compra = int(input())
    Productos.append(Prodto)
    Cantidad.append(Compra)
    print("de",Prodto,"usted adquirio",Compra,"ahora tenemos:",Compra)

  print("Productos:",Productos)
  print("Cantidad:",Cantidad)
```
